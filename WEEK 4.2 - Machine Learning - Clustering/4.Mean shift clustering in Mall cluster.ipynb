```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
#iloc ==> is used to take columns as input
#  [:,:] = [full row , full column]  = [1,1] means row first & columns select 
#  [:, 3,4]  means full row & 3 & 4 coulnms will take as input
```

```python
X = dataset.iloc[:, [3, 4]].values   #if we given [3:5] means will take from cpoumn 3 to column 5 ex[1:10], [1:4]
X
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

# Estimate a more appropriate bandwidth
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=500)   #quantile=0.2 is 20th percent of all pairwise distance


# Apply MeanShift with the estimated bandwidth
clusmodel = MeanShift(bandwidth=bandwidth)
label = clusmodel.fit_predict(X)
label
```

```python
supervised=dataset
supervised
```

```python
supervised['Cluster_group']=label
supervised
```

```python
supervised.to_csv("MeanShift_Mall cluster.csv",index=False)
```

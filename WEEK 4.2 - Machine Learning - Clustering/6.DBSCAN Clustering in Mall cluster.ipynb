```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
#iloc ==> is used to take columns as input
#  [:,:] = [full row , full column]  = [1,1] means row first & columns select 
#  [:, 3,4]  means full row & 3 & 4 coulnms will take as input
```

```python
X = dataset.iloc[:, [3, 4]].values   #if we given [3:5] means will take from cpoumn 3 to column 5 ex[1:10], [1:4]
X
```

```python
from sklearn.cluster import DBSCAN

clusmodel = DBSCAN(eps=0.5, min_samples=10)    #eps=- maximum distance between two samples
label = clusmodel.fit_predict(X)
label
```

```python
#in previous step eps is very low & sample is high compare of the distance
from sklearn.cluster import DBSCAN

clusmodel = DBSCAN(eps=2, min_samples=10)    #eps=- maximum distance between two samples
label = clusmodel.fit_predict(X)
label
```

```python
#in previous step eps is very low & sample is high compare of the distance
from sklearn.cluster import DBSCAN

clusmodel = DBSCAN(eps=2, min_samples=5)    #eps=- maximum distance between two samples
label = clusmodel.fit_predict(X)
label
```

```python
#in previous step eps is very low & sample is high compare of the distance
from sklearn.cluster import DBSCAN

clusmodel = DBSCAN(eps=0.2, min_samples=1)    #eps=- maximum distance between two samples
label = clusmodel.fit_predict(X)
label
```

```python

from sklearn.cluster import DBSCAN

clusmodel = DBSCAN(eps=0.5, min_samples=1)    #eps=- maximum distance between two samples
label = clusmodel.fit_predict(X)
label
```

```python
supervised=dataset
supervised
```

```python
supervised['Cluster_group']=label
supervised
```

```python
supervised.to_csv("DBSCAN_Mall cluster.csv",index=False)
```
